Pandas es una herramienta que nos permite manipular sets de datos

Para la etapa de pre-procesado de datos, según sea el caso, existen varias operaciones que querríamos efectuar:
    . Adecuar los datos para facilitar su visualización en un análisis exploratorio
    . Fusionar varios dataset
    . Reemplazar/rellenar/corregir/eliminar datos
    . Cambiar el formato de los datos
    . Reducción de dimensiones
    
Para ejemplificar el uso de algunas operaciones de _pandas_ se utilizará el conjunto de datos de dos dimensiones "heart-disease-uci", el cual describe los atributos de pacientes de distintos rangos etarios, así como la presencia de algún padecimiento cardíaco (https://www.kaggle.com/ronitf/heart-disease-uci).

```python
# se asignan estos alias simplemente por convención
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart.csv', delimiter = ',')
```

Podemos consultar atributos con valores específicos. Por ejemplo, los pacientes que tienen 57 años:

```python
# al utilizar los paréntesis cuadrados, pandas evaluará las expresiones que contenga
# se evalúa la expresión para cada fila, filtrando así aquellas en las cuales no se cumpla
df[df.age == 57]
```

Podemos también ordenar los datos según un criterio, lo cual, como será evidenciado, desordena los índices originales de cada entrada:

```python
# ordenando por valores de colesterol en la sangre, de manera ascendente
df.sort_values('chol')
```

```python
# ordenando por edad y niveles de colesterol (nótese los índices desordenados)
df.sort_values(by = ['age', 'chol'])
```

Podemos solicitar la re-asignación de índices, de manera que no interfieran con la visualización de los datos:

```python
df.sort_values(by = ['age'], ascending = False).reset_index()
```

Existen conjuntos de datos con valores desconocidos y/o en un formato inadecuado. Para afrontarles podemos utilizar las funciones _replace_, _dropna_, _fillna_ o una combinación:

```python
# supongamos que el dataset tiene caracteres '?' para algunos de sus atributos, 
# podemos convertirlos en lo que queramos, en este caso NaN
df = df.replace('?', np.nan)

# podemos así rellenar esos NaN con valores a gusto (fillna), o removerlos completamente
# en este caso rellenamos con ceros, podemos rellenar con valores estadísticos como se ve en las líneas comentadas
df.fillna(0)
# df.fillna(df.mean())
# df['chol'].fillna(df['chol'].median())

# el parámetro how acepta 'all' o 'any'. 'all' elimina la fila/columna sii TODOS sus valores son NaN, 
# mientras que 'any' lo hace si cualquiera de sus valores es NaN
# en axis, 0 = 'index' : filas. 1 = 'columns' : columnas
df = df.dropna(axis=0, how="any")
```

Podemos también eliminar features a gusto (para este conjunto de datos, es necesario indicar en el read_csv que la primer fila contiene los índices de las columnas)

```python
df = pd.read_csv('heart.csv', delimiter = ',', index_col=0)

# lista de features para eliminar
to_eliminate = ['restecg', 'exang']

# drop según columnas (axis=1)
df = df.drop(labels=['sex'], axis=1)

# drop utilizando parámetro 'inplace', de manera que se efectúe directamente sobre el dataset
# y no en su valor de retorno: sería equivalente a df = df.drop()
df.drop(labels=to_eliminate, axis=1, inplace=True)

df
```

Podemos también dar estilo al feature categórico de este set de datos, para que en su visualización(PCA) más adelante sea más facil discriminar los pacientes con pacedimientos cardíacos de los que no tienen.

```python
targets = df['target'].astype('category')
label_color = ['red' if i==1 else 'green' for i in targets]
```

# Análisis exploratorio

```python
# reestablecemos el set de datos
df = pd.read_csv('heart.csv', delimiter = ',')
```

**Descripción general de los datos**

```python
df.describe()
```

**Gráficas descriptivas de algunos features**

```python
import matplotlib.pyplot as plt 

fig, graph = plt.subplots()

mu = df['age'].mean()  # media de la distribución
sigma = df['age'].std()  # desviación estándar de la distribución
num_bins = 50

# histograma de distribución etaria
n, bins, patches = graph.hist(df['age'], num_bins, density=1)

y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

graph.plot(bins, y, '--')
graph.set_xlabel('Edad')
graph.set_ylabel('Densidad')
graph.set_title(r'Histograma de distribución etaria')

plt.show()
```

```python
df1 = df.groupby(['sex', 'target'])[['sex', 'target']]
df2 = pd.DataFrame(df1.size()).reset_index()
df2.columns = ['sex', 'target', 'count']


y_pos = np.arange(len(df2['target']))

plt.bar(y_pos, df2['count'], align='center', alpha=0.5)

plt.xticks(y_pos, ('M-Padecen', 'M-Sanas', 'H-Padecen', 'H-Sanos'))
plt.ylabel('Cantidad de personas')
plt.title('Incidencia de padecimientos cardíacos según género')

plt.show()

df2
```

**PCA**

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df)
new_data = pca.transform(df)
# debemos cambiar 'new_data' a un objeto DataFrame de Pandas
new_data = pd.DataFrame(new_data)
```

Procedemos con el ploteo de los datos

```python
new_data.columns = ['PCA component 1', 'PCA component 2']
new_data.plot.scatter(x='PCA component 1', y='PCA component 2', marker='o',
        alpha=0.7, # opacidad
        color=label_color, # lo definimos más arriba, colores verde y rojo
        title="Rojo: padece del corazón, Verde: no padece")
plt.show()
```
